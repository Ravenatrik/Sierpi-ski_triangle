```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(0,0,s=1)
plt.scatter(4,0,s=1)
plt.scatter(2,np.sqrt(12),s=1)
plt.scatter(2.5,1.5,s=1)

x=2.5
y=1.5
X=np.array([0,4,2])
Y=np.array([0,0,np.sqrt(12)])
xarr = []
yarr = []

for i in range(10000):
    j=np.random.randint(3)
    x=(x+X[j])/2
    y=(y+Y[j])/2
    xarr.append(x)
    yarr.append(y)

plt.scatter(xarr,yarr,s=1)




plt.show()
```
